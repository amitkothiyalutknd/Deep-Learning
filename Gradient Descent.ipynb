```python
def sigmoid(x):
    import math
    return 1/(1 + math.exp(-x))
sigmoid(86)
```

```python
coef1, coef2, coef3, intercept = -0.17884584, 0.9970676, 1.6462048, 0.81756586
def prediction_function(Percentage, ResumeScore, InterviewScore):
    weigthedsum = coef1 * Percentage + coef2 * ResumeScore + coef3 * InterviewScore + intercept
    return sigmoid(weigthedsum)
```

```python
def log_loss(ytrue, ypredicted):
    epsilon = 1e-15
    ypredictednew = [max(i, epsilon) for i in ypredicted]
    ypredictednew = [max(i, epsilon) for i in ypredicted]
    ypredictednew = ny.array(ypredictednew)
    return -ny.mean(ytrue * ny.log(ypredictednew) + (1- ytrue) * ny.log(1 - ypredictednew))
```

```python
def sigmoid_numpy(x):
    return 1/(1 + ny.exp(-x))
sigmoid_numpy(ny.array([16, 1, 0]))
```

```python
def gradient_descent(Percentage, ResumeScore, InterviewScore, ytrue, epochs, lossthresold):
    w1 = w2 = w3 = 1
    bias = 0
    rate = 0.5
    n = len(Percentage)
    
    for i in range(epochs):
        weighted_sum = w1 * Percentage + w2 * ResumeScore + w3 * InterviewScore + bias
        ypredicted = sigmoid_numpy(weighted_sum)
        
        loss = log_loss(ytrue, ypredicted)
        w1d = (1/n) * ny.dot(ny.transpose(Percentage), (ypredicted - ytrue))
        w2d = (1/n) * ny.dot(ny.transpose(ResumeScore), (ypredicted - ytrue))
        w3d = (1/n) * ny.dot(ny.transpose(InterviewScore), (ypredicted - ytrue))
        
        bias_d =  ny.mean(ypredicted - ytrue)
        w1 = w1 - rate * w1d
        w2 = w2 - rate * w2d
        w3 = w3 - rate * w3d
        bias = bias - rate *  bias_d
        
        print(f'Epoch:{i}, w1:{w1}, w2:{w2}, w3:{w3}, bias:{bias}, loss:{loss}')
        if loss <= lossthresold:
            break
        
    return w1, w2, w3, bias
```

```python
class myNN:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.w3 = 1
        self.bias = 0
        
    def  fit(self, x, y, epochs, lossthresold):
        self.w1, self.w2, self.w3, self.bias = self.gradient_descent(x['Percentage'], x['ResumeScore'],x['InterviewScore'], y, epochs, lossthresold)

    def predict(self, xtest):
        weighted_sum = self.w1 * xtest['Percentage'] + self.w2 * xtest['ResumeScore'] + self.w3 * xtest['InterviewScore'] + self.bias
        return sigmoid_numpy(weighted_sum)
    
    def gradient_descent(self, Percentage, ResumeScore, InterviewScore, ytrue, epochs, lossthresold):
        w1 = w2 = w3 = 1
        bias = 0
        rate = 0.5
        n = len(Percentage)
    
        for i in range(epochs):
            weighted_sum = w1 * Percentage + w2 * ResumeScore + w3 * InterviewScore + bias
            ypredicted = sigmoid_numpy(weighted_sum)
        
            loss = log_loss(ytrue, ypredicted)
            w1d = (1/n) * ny.dot(ny.transpose(Percentage), (ypredicted - ytrue))
            w2d = (1/n) * ny.dot(ny.transpose(ResumeScore), (ypredicted - ytrue))
            w3d = (1/n) * ny.dot(ny.transpose(InterviewScore), (ypredicted - ytrue))
        
            bias_d =  ny.mean(ypredicted - ytrue)
            w1 = w1 - rate * w1d
            w2 = w2 - rate * w2d
            w3 = w3 - rate * w3d
            bias = bias - rate *  bias_d
        
            if i%50 == 0:
                print(f'Epoch:{i}, w1:{w1}, w2:{w2}, w3:{w3}, bias:{bias}, loss:{loss}')
        
            if loss <= lossthresold:
                print(f'Epoch:{i}, w1:{w1}, w2:{w2}, w3:{w3}, bias:{bias}, loss:{loss}')
                break
#             sys.exit()

        return w1, w2, w3, bias
```
