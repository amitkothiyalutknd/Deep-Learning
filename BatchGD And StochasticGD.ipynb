```python
def prediction(area, bedroom, w, b):
    xscaled = sx.transform([[area, bedroom]])[0]
    scaledPrice = w[0] * xscaled[0] + w[1] * xscaled[1] + b
    return sy.inverse_transform([[scaledPrice]])[0][0]

prediction(2087, 2, w, b)
```

```python
import random
def stochasticgradeientDescent(x, ytrue, epochs, learningrate = 0.01):
    numberoffeatures = x.shape[1]
    w = ny.ones(shape = (numberoffeatures))
    b = 0
    totalsamples = x.shape[0]
    
    costList = []
    epochList = []
    
    for i in range(epochs):
        randomIndex = random.randint(0, totalsamples - 1)
        xsample = x[randomIndex]
        ysample = ytrue[randomIndex]# w1 * area + w2 * bedrooms
        yprediction = ny.dot(w, xsample.T) + b
        
        wgrad = -(2/totalsamples) * (xsample.T.dot(ysample -  yprediction))
        bgrad = -(2/totalsamples) * (ysample - yprediction)
        
        w = w - learningrate * wgrad
        b = b - learningrate * bgrad
        
        cost = ny.square(ysample - yprediction)
        if i%10 == 0:
            costList.append(cost)
            epochList.append(i)
    return w, b, cost, costList, epochList

wsgd, bsgd, costsgd, costListsgd, epochListsgd = stochasticgradeientDescent(xscaled, yscaled.reshape(yscaled.shape[0], ), 500)
wsgd, bsgd, costsgd
```
