```python
import time

def fib1(n):
    if n == 0 or n == 1:
        return 1
    else:
        return fib1(n - 1) + fib1(n - 2)

fib1(40)
start1 = time.time()
end1 = time.time() - start1
print(fib1(40),"\t", end1)
```

```python
def fib2(n, d):
    if n in d:
        return d[n]
    else:
        d[n] = fib2(n -1, d) + fib2(n - 2, d)
        return d[n]

d = {0:1,1:1}    
fib2(40, d)
start2 = time.time()
end2 = time.time() - start2
print(fib2(40, d),"\t",  end2)
```
