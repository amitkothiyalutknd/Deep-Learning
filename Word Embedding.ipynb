```python
reviews = ['Nice Tone', 'Delicious Taste', 'Uprising the growth.', 'Just Loved It!', 'Improve Performance', 'Nice Movie', 'Scaring Movies', 'Dangerous Site', 'Poor Network', 'Consistency', 'Keep it up']
reactions = ny.array([1, 1, 1, 1, 1, 1, 0, 0, 0, 1, 0 ])
```

```python
one_hot('Improve Performance', 110)
```

```python
vocab_size = 50
encoded_reviews = [one_hot(d, vocab_size) for d in reviews]
encoded_reviews
```

```python
max_length = 6
padded_reviews = pad_sequences(encoded_reviews, maxlen = max_length, padding = 'post')
padded_reviews
```

```python
embeded_vector_size = 7

model = Sequential()
model.add(Embedding (vocab_size, embeded_vector_size, input_length = max_length, name = 'embedding'))
model.add(Flatten())
model.add(Dense(1, activation = 'sigmoid'))
```

```python
x = padded_reviews
y = reactions
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

```python
model.fit(x, y, epochs = 50, verbose = True)
```

```python
loss, accuracy = model.evaluate(x, y)
```
