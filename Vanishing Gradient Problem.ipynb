```python
x, y = make_moons(n_samples = 250, noise = 0.05, random_state = 42)
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y, s = 100)
plt.show()
```

```python
model1 = Sequential()

model1.add(Dense(10, activation = 'sigmoid', input_dim = 2))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(10, activation = 'sigmoid'))
model1.add(Dense(1, activation = 'sigmoid'))
```

```python
model1.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model1.get_weights()[0]
```

```python
oldweights1 = model1.get_weights()[0]
oldweights1
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
model1.fit(xtrain, ytrain, epochs = 1)
```

```python
newweights1 = model1.get_weights()[0]
newweights1
```

```python
model1.optimizer.get_config()["learning_rate"]
```

```python
gradient1 = (oldweights1 - newweights1)/ 0.001
percentagechange1 = abs(100 * (oldweights1 - newweights1)/oldweights1)
```

```python
gradient1
```

```python
percentagechange1
```
