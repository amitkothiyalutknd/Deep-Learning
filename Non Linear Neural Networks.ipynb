```python
from keras.models import Model
from keras.layers import *
```

```python
x = Input(shape = (3, ))
```

```python
hiddenLayer1 = Dense(128, activation = 'relu')(x)
hiddenLayer2 = Dense(64, activation = 'relu')(hiddenLayer1)

outputLayer1 = Dense(1, activation = 'linear')(hiddenLayer2)
outputLayer2 = Dense(1, activation = 'sigmoid')(hiddenLayer2)

#One Input Layer and Multiput Layer Outputs
model = Model(inputs = x, outputs = [outputLayer1, outputLayer2])

model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```

```python
plot_model(model, show_shapes = True)
```

```python
from keras.layers import *
from keras.models import Model
```

```python
#Two sets of input layers
inputA = Input(shape = (32, ))
inputB = Input(shape = (128, ))
```

```python
#The First branch operates on the first input
x = Dense(8, activation = 'relu')(inputA)
x1 = Dense(4, activation = 'relu')(x)
```

```python
#The Second branch operates on the second input
y = Dense(64, activation = 'relu')(inputB)
y1 = Dense(32, activation = 'relu')(y)
y2 = Dense(4, activation = 'relu')(y1)
```

```python
#Merge the outputs of the two brances
combined = concatenate([x1, y2])
```

```python
#Apply a FC layer and then a regression prediction on the combined outputs
z = Dense(2, activation = 'relu')(combined)
z1 = Dense(1, activation = 'linear')(z)
```

```python
#Out Model will accept the inputs of the two brances and then output a single value
model = Model(inputs = [inputA, inputB], outputs = z)
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes = True)
```
